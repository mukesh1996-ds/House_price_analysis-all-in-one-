```python
# checcking the null values
df.isnull().sum().head(40) , df.isnull().sum().tail(41)
```

From the above we can see that there are high missing value and also in multiple columns. Let's deal with it first.

```python
# calculating the percentage of missing value 
percentage_missing = df.isnull().sum() * 100/len(df)
percentage_missing
```

```python
# creating a dataframe
#missing_value_df = pd. DataFrame({'percentage_missing': percentage_missing})
```

```python
#missing_value_df
```

```python
# Getting the missing value with more that 20 % of missing value
missing_value_20 = percentage_missing[percentage_missing > 20].keys()
missing_value_20
```

```python
# Final drop the missing value
new_df = df.drop(columns=missing_value_20)
new_df.shape
```

From the above step we  have drop the missing values with height percentages

```python
# creating a new data frame which include int and float values only
new_df.dropna(inplace=True)
```

```python
# rechecking the missing value
new_df.isnull().sum()
```

From the above we can seee that there are only 2 columns left with missing value that with int and float let's replace it .

```python
# removing the idcolumns 
new_df.drop(columns=['Id'], axis=1, inplace=True)
```
